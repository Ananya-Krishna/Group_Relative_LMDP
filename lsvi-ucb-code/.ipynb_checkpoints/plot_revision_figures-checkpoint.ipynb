We want to compare the following methods:
    
    [1] LSVI-UCB

    [2] RLSVI 
    
    [3] adaptive epsilon-greedy


**Settings**:

[a]. To see the regret does not scale with S, we try different S --- S = 100, 200, 400. 

[b]. We also try the same setting with misspecification probability = 0.05

```python
# aggregate the results 


def Regret_aggregation(path_to_data, num_epi, num_trials):
    '''

    Aggregate the regret results from data
    
    inputs:
    path_to_data: path to the files 
    num_epi: number of episodes
    num_trials: number of experiments -- ID
    
    '''
    
    ID = range(1, num_trials+1)
    S_seq = [100, 200, 400]
    Mis = [0, 50]
    
    # regret_eps = np.zeros( 3, 2, num_epi)
    regret_adeps_final = np.zeros((3, 2, num_epi))
    regret_ucb_final = np.zeros((3, 2, num_epi))
    regret_rlsvi1_final = np.zeros((3, 2, num_epi))
    regret_rlsvi2_final = np.zeros((3, 2, num_epi))
    
    for S in range(0,3):
        for mis in range(0,2):
            mean_adeps = np.zeros(num_epi,)
            mean_ucb = np.zeros(num_epi,)
            mean_rlsvi1 = np.zeros(num_epi,)
            mean_rlsvi2 = np.zeros(num_epi,)
            max_return = 0
            for id_number in ID:
                data = np.load(path_to_data + 'result_s'+str(S_seq[S])+'_a5_h10_d10_mis'+str(Mis[mis])+'_id'+str(id_number)+'.npz')

                # print("datasets in the file are", data.files)
                # eps = data['episode_accu_result_eps'] # epsilon greedy 
                adeps = data['episode_accu_result_adeps']
                ucb = data['episode_accu_result_ucb'] # LSVI-UCB
                optimal = data['episode_accu_result_opt'] # optimal policy 
                rlsvi1 = data['episode_accu_result_lsvi1']
                rlsvi2 = data['episode_accu_result_lsvi2']

                # mean_eps = np.asarray(eps).mean(axis= 0)
                mean_adeps = mean_adeps + np.asarray(adeps).mean(axis=0)
                mean_ucb = mean_ucb + np.asarray(ucb).mean(axis=0)
                mean_rlsvi1= mean_rlsvi1 + np.asarray(rlsvi1).mean(axis=0)
                mean_rlsvi2 = mean_rlsvi2 + np.asarray(rlsvi2).mean(axis=0)

                #print(np.shape(result_s100_id1["episode_accu_result_eps"]))
                # ucb = result_s100_id1['episode_accu_result_ucb']

                max_return = np.maximum(np.max(optimal),max_return)
            
            # finish processing each ID's 
            
            # regret benchmark = maximum_mean_reward * sequence
            max_reward = max_return / num_epi 
            regret_benchmark = np.asarray(range(1, num_epi + 1)) * max_reward 


            regret_adeps = regret_benchmark - mean_adeps/num_trials
            regret_ucb = regret_benchmark - mean_ucb/num_trials
            regret_rlsvi1 = regret_benchmark - mean_rlsvi1/num_trials
            regret_rlsvi2 = regret_benchmark - mean_rlsvi2/num_trials
            
            num_plot = 3000
            hori_plot = range(1, num_plot + 1)
            plt.figure(figsize=(8,6))
            y = range(1, num_plot + 1)
            # plt.plot(y, eps,  label='Epsilon-Greedy')
            plt.plot(y, regret_adeps[range(0, num_plot)], label='Epsilon-Greedy')
            plt.plot(y, regret_ucb[range(0, num_plot)], label='LSVI-UCB')
            plt.plot(y, regret_rls1[range(0, num_plot)], label='LSVI-RLSVI-1')
            plt.plot(y, regret_rls2[range(0, num_plot)], label='LSVI-RLSVI-2')
            plt.xlabel('Episode')
            plt.ylabel('Regret')
            plt.legend()
            plt.title('regret_s'+str(S_seq[S])+'_mis'+str(Mis[mis]))
            
            
            # store results
            
            regret_adeps_final[S,mis,:] = regret_adeps 
            regret_ucb_final[S,mis,:] = regret_ucb
            regret_rlsvi1_final[S,mis,:] = regret_rlsvi1
            regret_rlsvi2_final[S,mis,:] = regret_rlsvi2
            
    file_name = "regret_results_for_plot.npz"
    
    np.savez(file_name, num_epi = num_epi, regret_adeps = np.asarray(regret_adeps_final), regret_ucb = np.asarray(regret_ucb_final), 
        regret_rlsvi1 = np.asarray(regret_rlsvi1_final), regret_rlsvi2 = np.asarray(regret_rlsvi2_final))
    
    
    

    
def UCB_aggregation(path_to_data, num_epi, num_trials):
    '''
    Aggregate the regret results from data
    
    inputs:
    path_to_data: path to the files 
    num_epi: number of episodes
    num_trials: number of experiments -- ID
    
    '''
    
    ID = range(1, num_trials+1)
    S_seq = [100, 200, 400]
    Mis = [0, 50]
    
    
    regret_ucb_final = np.zeros((3, 2, num_epi))
 
    for S in range(0,3):
        for mis in range(0,2):
            print("handle S "+str(S_seq[S]) + " mis "+str(Mis[mis]))
            ucb_rewards = np.zeros(num_epi,)
 
            max_return = 0
            count = 0
        
            for id_number in ID:
                data = np.load(path_to_data + 'ucb_result_s'+str(S_seq[S])+'_a5_h10_d10_mis'+str(Mis[mis])+'_id'+str(id_number)+'.npz')
                 
                ucb = data['episode_accu_result_ucb'] # LSVI-UCB
                optimal = data['episode_accu_result_opt'] # optimal policy
                num_repeat = np.shape(ucb)[0]
                #print(np.shape(ucb))
                #print(ucb[-1])
                average_ucb = ucb[:,-1] / num_epi
                valid_index = average_ucb > 0.1
                # we need to remove the episodes where the methods fails 
                for i in range(num_repeat):
                    if valid_index[i]:
                        ucb_rewards = ucb_rewards + ucb[i,:]
                        count = count + 1
              
                max_return = np.maximum(np.max(optimal),max_return)
            
            # finish processing each ID's 
            
            # regret benchmark = maximum_mean_reward * sequence
            max_reward = max_return / num_epi 
            
            # print(ucb_rewards[-1]/(num_epi*count), max_reward)
            
            regret_benchmark = np.asarray(range(1, num_epi + 1)) * max_reward 
 
            regret_ucb = regret_benchmark - ucb_rewards / count

    
            num_plot = 3000
            hori_plot = range(1, num_plot + 1)
            plt.figure(figsize=(8,6))
            y = range(1, num_plot + 1)
            # plt.plot(y, eps,  label='Epsilon-Greedy')
            # plt.plot(y, regret_adeps[range(0, num_plot)], label='Epsilon-Greedy')
            plt.plot(y, regret_ucb[range(0, num_plot)], label='LSVI-UCB')
        
            plt.xlabel('Episode')
            plt.ylabel('Regret')
            plt.legend()
            plt.title('UCB_regret_s'+str(S_seq[S])+'_mis'+str(Mis[mis]))
            
            
 
            regret_ucb_final[S,mis,:] = regret_ucb
 
    file_name = "ucb_regret_results_for_plot.npz"
    
    np.savez(file_name, num_epi = num_epi, regret_ucb = np.asarray(regret_ucb_final))
    
    
```

#### Generate the plots 




```python
%matplotlib inline
import matplotlib as mpl
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import UnivariateSpline

plt.rcParams['font.size'] = 25


Regret_Data = np.load("final_regret_result.npz")

num_epi = Regret_Data["num_epi"]
regret_ucb = Regret_Data["regret_ucb"]
regret_eps = Regret_Data["regret_adeps"]
regret_rlsvi1 = Regret_Data["regret_rlsvi1"]
regret_rlsvi2 = Regret_Data["regret_rlsvi2"]


S_seq = [100, 200, 400]
Mis = [0, 50]

smoothing_parameter = 0.001


## first generate the plots for each stepup
for S in range(3):
    for mis in range(2):
        ucb = regret_ucb[S, mis, : ]
        rlsvi = regret_rlsvi2[S, mis, : ]
        eps = regret_eps[S, mis, :]
        
        x = range(1, num_epi+1)
        
        spl_ucb = UnivariateSpline(x, ucb, s = smoothing_parameter)
        spl_rlsvi = UnivariateSpline(x, rlsvi, s = smoothing_parameter)
        spl_eps = UnivariateSpline(x, eps, s = smoothing_parameter)
        
        xs = np.linspace(10, num_epi, 10000)
        ucb_smooth = spl_ucb(xs)
        rlsvi_smooth = spl_rlsvi(xs)
        eps_smooth = spl_eps(xs)
        
        
        figure = plt.figure(figsize=(12,9))
        # plot these data 

        plt.plot(xs, eps_smooth, color='crimson', linewidth=4, linestyle = '-', label='Epsilon-Greedy')
        plt.plot(xs, rlsvi_smooth, color='royalblue', linewidth=5, linestyle = '--', label='RLSVI')
        plt.plot(xs, ucb_smooth, color='darkgreen', linewidth=5, linestyle = '-.', label='LSVI-UCB')

        
        plt.legend(frameon=False)

        plt.savefig('./paper_plots/S'+str(S_seq[S])+'mis'+str(Mis[mis])+'_nolabel.pdf', dpi=800, transparent=True)

        plt.xlabel('Episodes', loc = 'right')
        plt.ylabel('Regret', loc='top')
        plt.savefig('./paper_plots/S'+str(S_seq[S])+"mis"+str(Mis[mis])+'.pdf', dpi=800, transparent=True, bbox_inches='tight')
        plt.show()
        
        
        # only rlsvi_vs_ucb
        figure = plt.figure(figsize=(12,9))
        # plot these data 

        plt.plot(xs, rlsvi_smooth, color='royalblue', linewidth=5, linestyle = '--', label='RLSVI')
        plt.plot(xs, ucb_smooth, color='darkgreen', linewidth=5, linestyle = '-.', label='LSVI-UCB')

        
        plt.legend(frameon=False)

        plt.savefig('./paper_plots/S'+str(S_seq[S])+'mis'+str(Mis[mis])+'v2_nolabel.pdf', dpi=800, transparent=True)

        plt.xlabel('Episodes', loc = 'right')
        plt.ylabel('Regret', loc='top')
        plt.savefig('./paper_plots/S'+str(S_seq[S])+"mis"+str(Mis[mis])+'v2.pdf', dpi=800, transparent=True, bbox_inches='tight')
        plt.show()
                                        




                
                                        
```
