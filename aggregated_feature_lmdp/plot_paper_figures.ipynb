```python
# process the data --- return regret & normalized regret

def process_data(Name, normalize = False):
    
    """
    do some preprocessing on the dataset
    input: 
        name of the dataset 
    output: (y1,y2) 
    y1 = epsilon-greedy
    y2 = lsvi 
    
    """
    
    data = np.load(Name)
    # print("datasets in the file are", data.files)
    
    eps = data['episode_accu_result_eps'] # epsilon greedy 
    ucb = data['episode_accu_result_ucb'] # LSVI-UCB
    optimal = data['episode_accu_result_opt'] # optimal policy 

    # print(Name, "the shape of the dataset is", eps.shape) 
 

    n_trial = eps.shape[0]
    n_episode = eps.shape[1]


    max_reward = np.max(optimal) / n_episode 

    
    # print(Name, "episode return of the optimal policy is ", max_reward)
    regret_benchmark = np.asarray(range(1, n_episode + 1)) * max_reward 

    # mean of cummulative episode return 
    mean_eps = np.asarray(eps).mean(axis=0)
    mean_ucb = np.asarray(ucb).mean(axis=0)
    
    # regret 
    regret_eps = regret_benchmark - mean_eps
    regret_ucb = regret_benchmark - mean_ucb
    
    if normalize == True:
        regret_eps = np.divide(regret_eps, max_reward)
        regret_ucb = np.divide(regret_ucb, max_reward)
    
    return regret_eps, regret_ucb

 
```
